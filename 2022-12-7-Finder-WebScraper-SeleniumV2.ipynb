```python
import datetime
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
url = 'https://www.finder.com.au/home-loans'
driver.get(url)

driver.execute_script("window.scrollTo(0, 2080)") 

get_today = datetime.datetime.now()
today = get_today.strftime('%d/%m/%Y')
affiliate = 'Finder'
rank = 1

results = [['Date', 'Affiliate', 'Position', 'Provider', 'Product', 'Interest Rate', 'Comparison Rate', 'Monthly Repayment', 'Offer']]

loans = WebDriverWait(driver, 30).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="comparison-table-0000000000"]/tbody//tr')))

for i in range(1, len(loans)):
    loan_listing = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="comparison-table-0000000000"]/tbody/tr[{i}]//a[1]'))).text.split(' ', 1)
    loan_interest_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="comparison-table-0000000000"]/tbody/tr[{i}]/td[2]/div/div[2]/span'))).text
    loan_comparison_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="comparison-table-0000000000"]/tbody/tr[{i}]/td[3]/div/span[1]'))).text
    loan_monthly_repayment = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="comparison-table-0000000000"]/tbody/tr[{i}]/td[5]/div/span'))).text
    
    loan_offer = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="comparison-table-0000000000"]/tbody/tr[{i}]/td[6]'))).text.replace('\n', ' ')
    if 'Go to site' in loan_offer :
        loan_entry = str(today) + '|' + str(affiliate) + '|' + str(rank) + '|' + str(loan_listing[0]) + '|' + str(loan_listing[0]) + ' ' + str(loan_listing[1]) + '|' + str(loan_interest_rate) + '|' + str(loan_comparison_rate) + '|' + str(loan_monthly_repayment) + '|' + str('No offer')
    else :
        loan_entry = str(today) + '|' + str(affiliate) + '|' + str(rank) + '|' + str(loan_listing[0]) + '|' + str(loan_listing[0]) + ' ' + str(loan_listing[1]) + '|' + str(loan_interest_rate) + '|' + str(loan_comparison_rate) + '|' + str(loan_monthly_repayment) + '|' + str(loan_offer)
    
    rank = rank + 1
    results.append(loan_entry.split('|'))
    
df = pd.DataFrame(results[1:], columns=results[0]).to_dict('list')
df2 = pd.DataFrame(df)

print(df2)
df2.to_csv('finder.csv', index=False, encoding='utf-8')
```
