```python
import time
import datetime
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
url = 'https://www.canstar.com.au/home-loans'
driver.get(url)

time.sleep(2.5)

driver.execute_script("window.scrollTo(0, 2080)")

time.sleep(2.5)

driver.execute_script("window.scrollTo(0, 1080)")

get_today = datetime.datetime.now()
today = get_today.strftime('%d/%m/%Y')
affiliate = 'Canstar'
rank = 1

results = [['Date', 'Affiliate', 'Position', 'Provider', 'Product', 'Interest Rate', 'Comparison Rate', 'Monthly Repayment', 'Offer']]

loans = WebDriverWait(driver, 30).until(EC.presence_of_all_elements_located((By.XPATH, "//cnslib-table-card")))

for i in range(1, len(loans)):
    loan_listing =  WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f"((//cnslib-table-card)[{i}]//a)[1]"))).text.split(' | ')
    loan_interest_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f"((//cnslib-table-card)[{i}]/div/div/div/div[5]/div[1]/div/span)[1]"))).text
    loan_comparison_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f"((//cnslib-table-card)[{i}]/div/div/div/div[5]/div[2]/div/span)[1]"))).text
    loan_monthly_repayment = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f"((//cnslib-table-card)[{i}]/div/div/div/div[5]/div[3]/div/span)[1]"))).text
    loan_offer = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f"((//cnslib-table-card)[{i}]/div/div/div/div)[3]"))).text.replace('\n', ' ')
    
    if loan_offer == '' :
        loan_entry = (f'{today} | {affiliate} | {rank} | {loan_listing[0]} | {loan_listing[1]} {loan_listing[2]} | {loan_interest_rate} | {loan_comparison_rate} | {loan_monthly_repayment} | No offer')
    else :
        loan_entry = (f'{today} | {affiliate} | {rank} | {loan_listing[0]} | {loan_listing[1]} {loan_listing[2]} | {loan_interest_rate} | {loan_comparison_rate} | {loan_monthly_repayment} | {loan_offer}')

    rank = rank + 1

    results.append(loan_entry.strip().split(' | '))

df = pd.DataFrame(results[1:], columns=results[0]).to_dict('list')
df2 = pd.DataFrame(df)

print(df2)

today2 = get_today.strftime('%d-%m-%Y')
df2.to_csv(f'{today2}canstar.csv', index=False, encoding='utf-8')

```
