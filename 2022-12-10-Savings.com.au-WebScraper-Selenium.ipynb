```python
import time
import datetime
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
url = 'https://www.savings.com.au/home-loans/'
driver.get(url)

time.sleep(2.5)

driver.execute_script("window.scrollTo(0, 2080)")

time.sleep(2.5)

driver.execute_script("window.scrollTo(0, 1080)")

get_today = datetime.datetime.now()
today = get_today.strftime('%d/%m/%Y')
affiliate = 'Savings.com.au'
rank = 1

results = [['Date', 'Affiliate', 'Position', 'Provider', 'Product', 'Interest Rate', 'Comparison Rate', 'Monthly Repayment', 'Offer']]

loan_names = []
loan_products = []
loan_interest_rates = []
loan_comparison_rates = []
loan_monthly_repayments = []
loan_offers = []
results1 = [loan_names, loan_products, loan_interest_rates, loan_comparison_rates, loan_monthly_repayments, loan_offers]

loans = WebDriverWait(driver, 30).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody//tr[@class="product-data"]')))

for i in range(0, len(loans)) :   
    loan_name = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="brand-logo-{i}"]/img'))).get_attribute('alt')
    loan_names.append(loan_name)
    
for i in range(2, (len(loans) * 3), 3) :
    loan_product = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody/tr[{i}]//a'))).text
    loan_offer = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody/tr[{i}]/td[2]/div//ul[@class="feature-list"]'))).text
    loan_products.append(loan_product)
    loan_offers.append(loan_offer.replace('\n', ', '))
    
for i in range(1, (len(loans) * 3), 3) :
    loan_interest_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody/tr[{i}]/td[2]/app-sca-comparison-field/label//span'))).text
    loan_comparison_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody/tr[{i}]/td[3]/app-sca-comparison-field/label//span'))).text
    loan_monthly_repayment = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody/tr[{i}]/td[4]/app-sca-comparison-field/label//span'))).text
    loan_interest_rates.append(loan_interest_rate)
    loan_comparison_rates.append(loan_comparison_rate)
    loan_monthly_repayments.append(loan_monthly_repayment)
    
for i in range(0, len(loans)) :
    if results1[5][i] == '' :
        loan_entry = f'{today} | {affiliate} | {rank} | {results1[0][i]} | {results1[1][i]} | {results1[2][i]}% | {results1[3][i]}% | ${results1[4][i]} | No offer'
    else :
        loan_entry = f'{today} | {affiliate} | {rank} | {results1[0][i]} | {results1[1][i]} | {results1[2][i]}% | {results1[3][i]}% | ${results1[4][i]} | {results1[5][i]}'
    
    rank = rank + 1
    results.append(loan_entry.split('|'))

df = pd.DataFrame(results[1:], columns=results[0]).to_dict('list')
df2 = pd.DataFrame(df)

print(df2)
df2.to_csv('savingscomau.csv', index=False, encoding='utf-8')
```
