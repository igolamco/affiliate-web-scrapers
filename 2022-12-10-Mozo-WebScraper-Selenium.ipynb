```python
import datetime
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
url = 'https://mozo.com.au/top/home-loans'
driver.get(url)

driver.execute_script("window.scrollTo(0, 2080)") 

get_today = datetime.datetime.now()
today = get_today.strftime('%d/%m/%Y')
affiliate = 'Mozo'
rank = 1

results = [['Date', 'Affiliate', 'Position', 'Provider', 'Product', 'Interest Rate', 'Comparison Rate', 'Monthly Repayment', 'Offer']]

loans = WebDriverWait(driver, 30).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="compare"]/div[1]/div[2]/ul//li')))

for i in range(1, len(loans)):
    loan_name = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]//img'))).get_attribute('alt').split(' ')
    loan_product = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]//h6'))).text
    loan_product2 = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]//p'))).text
    
    loan_interest_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]/div[@class="b-mpp__data b-mpp__data--three"]/div[1]'))).text.replace('p.a.','').replace('\n','').replace('variable','')
    loan_comparison_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]/div[@class="b-mpp__data b-mpp__data--three"]/div[2]'))).text.replace('p.a.','')
    loan_monthly_repayment = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]/div[@class="b-mpp__data b-mpp__data--three"]/div[3]'))).text
    
    loan_offer = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]/p[@class="b-mpp__description"]'))).text
    
    loan_entry = f'{today} | {affiliate} | {rank} | {loan_name[-1]} | {loan_product} {loan_product2} | {loan_interest_rate} | {loan_comparison_rate} | {loan_monthly_repayment} | {loan_offer}'.strip()
    
    rank = rank + 1
    results.append(loan_entry.split('|'))

df = pd.DataFrame(results[1:], columns=results[0]).to_dict('list')
df2 = pd.DataFrame(df)

print(df2)
df2.to_csv('mozo.csv', index=False, encoding='utf-8')
```
