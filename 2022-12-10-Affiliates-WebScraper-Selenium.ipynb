```python
import time
import datetime
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
def finder_web_scrape(url, affiliate, wait_time_sec) :
    driver = webdriver.Chrome()
#     url = 'https://www.finder.com.au/home-loans'
    driver.get(url)

    time.sleep(wait_time_sec)

    driver.execute_script("window.scrollTo(0, 2080)")

    time.sleep(wait_time_sec)

    driver.execute_script("window.scrollTo(0, 1080)")

    get_today = datetime.datetime.now()
    today = get_today.strftime('%d/%m/%Y')
#     affiliate = 'Finder'
    rank = 1

#     results = [['Date', 'Affiliate', 'Position', 'Provider', 'Product', 'Interest Rate', 'Comparison Rate', 'Monthly Repayment', 'Offer']]

    loans = WebDriverWait(driver, 30).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="comparison-table-0000000000"]/tbody//tr')))

    for i in range(1, len(loans)):
        loan_listing = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="comparison-table-0000000000"]/tbody/tr[{i}]//a[1]'))).text.split(' ', 1)
        loan_interest_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="comparison-table-0000000000"]/tbody/tr[{i}]/td[2]/div/div[2]/span'))).text
        loan_comparison_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="comparison-table-0000000000"]/tbody/tr[{i}]/td[3]/div/span[1]'))).text
        loan_monthly_repayment = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="comparison-table-0000000000"]/tbody/tr[{i}]/td[5]/div/span'))).text
    
        loan_offer = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="comparison-table-0000000000"]/tbody/tr[{i}]/td[6]'))).text.replace('\n', ' ')
        if 'Go to site' in loan_offer :
            loan_entry = f'{today} | {affiliate} | {rank} | {loan_listing[0]} | {loan_listing[0]} {loan_listing[1]} | {loan_interest_rate} | {loan_comparison_rate} | {loan_monthly_repayment} | {"No offer"}'
        else :
            loan_entry = f'{today} | {affiliate} | {rank} | {loan_listing[0]} | {loan_listing[0]} {loan_listing[1]} | {loan_interest_rate} | {loan_comparison_rate} | {loan_monthly_repayment} | {loan_offer}'
    
        rank = rank + 1
        results.append(loan_entry.split('|'))

    driver.quit()
```

```python
def mozo_web_scrape(url, affiliate, wait_time_sec) :
    driver = webdriver.Chrome()
#     url = 'https://mozo.com.au/top/home-loans'
    driver.get(url)

    time.sleep(wait_time_sec)

    driver.execute_script("window.scrollTo(0, 2080)")

    time.sleep(wait_time_sec)

    driver.execute_script("window.scrollTo(0, 1080)")

    get_today = datetime.datetime.now()
    today = get_today.strftime('%d/%m/%Y')
#     affiliate = 'Mozo'
    rank = 1

#     results = [['Date', 'Affiliate', 'Position', 'Provider', 'Product', 'Interest Rate', 'Comparison Rate', 'Monthly Repayment', 'Offer']]

    loans = WebDriverWait(driver, 30).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="compare"]/div[1]/div[2]/ul//li')))

    for i in range(1, len(loans)):
        loan_name = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]//img'))).get_attribute('alt').split(' ')
        loan_product = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]//h6'))).text
        loan_product2 = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]//p'))).text
    
        loan_interest_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]/div[@class="b-mpp__data b-mpp__data--three"]/div[1]'))).text.replace('p.a.','').replace('\n','').replace('variable','')
        loan_comparison_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]/div[@class="b-mpp__data b-mpp__data--three"]/div[2]'))).text.replace('p.a.','')
        loan_monthly_repayment = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]/div[@class="b-mpp__data b-mpp__data--three"]/div[3]'))).text
    
        loan_offer = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="compare"]/div[1]/div[2]/ul/li[{i}]/p[@class="b-mpp__description"]'))).text
    
        if 'Qudos' in loan_name :
            loan_entry = f'{today} | {affiliate} | {rank} | {loan_name[-2]} | {loan_product} {loan_product2} | {loan_interest_rate} | {loan_comparison_rate} | {loan_monthly_repayment} | {loan_offer}'.strip()
        else :
            loan_entry = f'{today} | {affiliate} | {rank} | {loan_name[-1]} | {loan_product} {loan_product2} | {loan_interest_rate} | {loan_comparison_rate} | {loan_monthly_repayment} | {loan_offer}'.strip()
    
        rank = rank + 1
        results.append(loan_entry.split('|'))

    driver.quit()
```

```python
def ratecity_web_scrape(url, affiliate, wait_time_sec) :
    driver = webdriver.Chrome()
#     url = 'https://www.ratecity.com.au/home-loans'
    driver.get(url)

    time.sleep(wait_time_sec)

    driver.execute_script("window.scrollTo(0, 2080)")

    time.sleep(wait_time_sec)

    driver.execute_script("window.scrollTo(0, 1080)")

    get_today = datetime.datetime.now()
    today = get_today.strftime('%d/%m/%Y')
#     affiliate = 'RateCity'
    rank = 1

#     results = [['Date', 'Affiliate', 'Position', 'Provider', 'Product', 'Interest Rate', 'Comparison Rate', 'Monthly Repayment', 'Offer']]

    loans = WebDriverWait(driver, 30).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="rc-ratetable"]/div/div')))

    for i in range(1, len(loans)) :
        loan_name = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="rc-ratetable"]/div/div[{i}]//img'))).get_attribute('alt').split(' ')
        loan_product = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="rc-ratetable"]/div/div[{i}]//p'))).text
        loan_interest_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="rc-ratetable"]/div/div[{i}]/div[3]/div[2]/div[1]/div[2]/div[1]/div/div[1]/p[1]'))).text
        loan_comparison_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="rc-ratetable"]/div/div[{i}]/div[3]/div[2]/div[1]/div[2]/div[2]/div/div[1]/p[1]'))).text
        loan_monthly_repayment = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="rc-ratetable"]/div/div[{i}]/div[3]/div[2]/div[1]/div[2]/div[3]/div/p'))).text
        loan_offer = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="rc-ratetable"]/div/div[{i}]/div[2]/div[2]'))).text.replace('\n', '|').split('|')
    
        if 'Awards' in loan_offer[-1] or 'Extra repayments' in loan_offer[-1] or 'Offset Accounts' in loan_offer[-1] :
            loan_entry = f'{today} | {affiliate} | {rank} | {loan_name[0]} | {loan_product} | {loan_interest_rate} | {loan_comparison_rate} | {loan_monthly_repayment} | No offer'
        else:
            loan_entry = f'{today} | {affiliate} | {rank} | {loan_name[0]} | {loan_product} | {loan_interest_rate} | {loan_comparison_rate} | {loan_monthly_repayment} | {loan_offer[-2]}: {loan_offer[-1]}'
    
        rank = rank + 1
        results.append(loan_entry.split('|'))

    driver.quit()
```

```python
def savings_web_scrape(url, affiliate, wait_time_sec) :
    driver = webdriver.Chrome()
#     url = 'https://www.savings.com.au/home-loans/'
    driver.get(url)

    time.sleep(wait_time_sec)

    driver.execute_script("window.scrollTo(0, 2080)")

    time.sleep(wait_time_sec)

    driver.execute_script("window.scrollTo(0, 1080)")

    get_today = datetime.datetime.now()
    today = get_today.strftime('%d/%m/%Y')
#     affiliate = 'Savings.com.au'
    rank = 1

#     results = [['Date', 'Affiliate', 'Position', 'Provider', 'Product', 'Interest Rate', 'Comparison Rate', 'Monthly Repayment', 'Offer']]

    loan_names = []
    loan_products = []
    loan_interest_rates = []
    loan_comparison_rates = []
    loan_monthly_repayments = []
    loan_offers = []
    results1 = [loan_names, loan_products, loan_interest_rates, loan_comparison_rates, loan_monthly_repayments, loan_offers]

    loans = WebDriverWait(driver, 30).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody//tr[@class="product-data"]')))

    for i in range(0, len(loans)) :   
        loan_name = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="brand-logo-{i}"]/img'))).get_attribute('alt')
        loan_names.append(loan_name)
    
    for i in range(2, (len(loans) * 3), 3) :
        loan_product = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody/tr[{i}]//a'))).text
        loan_offer = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody/tr[{i}]/td[2]/div//ul[@class="feature-list"]'))).text
        loan_products.append(loan_product)
        loan_offers.append(loan_offer.replace('\n', ', '))
    
    for i in range(1, (len(loans) * 3), 3) :
        loan_interest_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody/tr[{i}]/td[2]/app-sca-comparison-field/label//span'))).text
        loan_comparison_rate = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody/tr[{i}]/td[3]/app-sca-comparison-field/label//span'))).text
        loan_monthly_repayment = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, f'//*[@id="sca-table-wrapper-61553d131a24daeba4e68036"]/table/tbody/tr[{i}]/td[4]/app-sca-comparison-field/label//span'))).text
        loan_interest_rates.append(loan_interest_rate)
        loan_comparison_rates.append(loan_comparison_rate)
        loan_monthly_repayments.append(loan_monthly_repayment)
    
    for i in range(0, len(loans)) :
        if results1[5][i] == '' :
            loan_entry = f'{today} | {affiliate} | {rank} | {results1[0][i]} | {results1[1][i]} | {results1[2][i]}% | {results1[3][i]}% | ${results1[4][i]} | No offer'
        else :
            loan_entry = f'{today} | {affiliate} | {rank} | {results1[0][i]} | {results1[1][i]} | {results1[2][i]}% | {results1[3][i]}% | ${results1[4][i]} | {results1[5][i]}'
    
        rank = rank + 1
        results.append(loan_entry.split('|'))

    driver.quit()
```

```python
bot_wait_time = 2.5
results = [['Date', 'Affiliate', 'Position', 'Provider', 'Product', 'Interest Rate', 'Comparison Rate', 'Monthly Repayment', 'Offer']]

get_today = datetime.datetime.now()
today = get_today.strftime('%d-%m-%Y')
    
finder_web_scrape('https://www.finder.com.au/home-loans', 'Finder', bot_wait_time)
mozo_web_scrape('https://mozo.com.au/top/home-loans', 'Mozo', bot_wait_time)
ratecity_web_scrape('https://www.ratecity.com.au/home-loans', 'RateCity', bot_wait_time)
savings_web_scrape('https://www.savings.com.au/home-loans/', 'Savings.com.au', bot_wait_time)

df = pd.DataFrame(results[1:], columns=results[0]).to_dict('list')
df2 = pd.DataFrame(df)

print(df2)
df2.to_csv(f'{today}affiliates_all.csv', index=False, encoding='utf-8')
```
